````python
def move_back(pos, steps, grid):
    row, col = pos
    lrows = len(grid)
    lcols = len(grid[0])
    total_steps = row*lcols+col
    new_pos_row, new_pos_col = divmod(total_steps - steps, lcols) 
    if 0<=new_pos_row<lrows and 0<=new_pos_col<lcols:
        return (new_pos_row, new_pos_col)
    else: return None
````
