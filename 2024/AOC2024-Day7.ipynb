```python
# updating find combinations to include '||'

def find_combinations_2(numbers_str):
    """
    Finds all possible combinations of operators (+ and *) for a given string of numbers.
    """
    key_operators = ['+', '*', '||']
    numbers = [int(x) for x in numbers_str.split()]
    num_operators = len(numbers) - 1
    all_combinations = []
    for operators_combination in itertools.product(key_operators, repeat=num_operators):
        expression = str(numbers[0])
        for i in range(num_operators):
            expression += f" {operators_combination[i]} {numbers[i+1]}"
        all_combinations.append(expression)
    return all_combinations

# updating eval function to include '||'
def evaluate_combination_left_to_right_2(combination):
    """Evaluates a mathematical expression string from left to right, ignoring precedence."""
    try:
        pts = combination.split("||")
        combination = "".join([p.strip() for p in pts])
        #print(combination)
        parts = combination.split()
        result = int(parts[0])
        for i in range(1, len(parts), 2):
            operator = parts[i]
            operand = int(parts[i+1])
            if operator == '+':
                result += operand
            elif operator == '*':
                result *= operand
        return result
    except (ValueError, IndexError):
        return None
```

```python
# Example usage with the new evaluation function:
numbers_str = "1 2 3 4"
combinations = find_combinations_2(numbers_str)

for combination in combinations:
    result = evaluate_combination_left_to_right_2(combination)
    print(f"{combination} = {result}")
```
