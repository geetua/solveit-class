````python
from aocd import get_data
inp = get_data(day=1, year=2024)
````

````python
def find_dist(locations):
    l1 = []
    l2 = []
    lines = locations.splitlines()
    for line in lines:
        (i, j) = line.split()
        l1.append(int(i))
        l2.append(int(j))
    l1.sort()
    l2.sort()
    total_dist = sum(abs(i-l2[index]) for index, i  in enumerate(l1))
    return total_dist
````

````python
total_dist = find_dist(locations)
total_dist
````

````python
total_dist = find_dist(inp)
total_dist
````

````python
def get_sim_score(locations):
    l1 = []
    l2 = defaultdict(int)
    lines = locations.splitlines()
    for line in lines:
        (i, j) = line.split()
        l1.append(int(i))
        l2[int(j)]+=1
    sim_score = sum(i*l2[i] for i in l1)
    return sim_score
````
